```
from pyspark.sql import SparkSession
from collections import namedtuple
```

```
class music_table:
    
    def __init__(self, appname, filetype, location, delimeter):
        self.spark = SparkSession.builder.appName(appname).getOrCreate()
        self.filetype = filetype
        self.location = location
        self.delimeter = delimeter
    
    def load_music_table(self):
        df = self.spark.read.format(self.filetype).option("inferSchema", "true").option("header", "true").option("sep", self.delimeter).load(self.location)
        return df
    
    def show_music_table(self):
        df.show()
    
    def printschema(self):
        df.printSchema()
        
    def get_music_table(self):
        return df
        
```

```
tracks_df = music_table(appname='name',location = "/FileStore/tables/tracks.csv", filetype = 'csv', delimeter = "	")
```

```
tracks_df.load_music_table()
```

```
tracks_df.show_music_table()
```

```
tracks_df.printschema()
```

```
tracks_df.get_music_table().show()
```
